# Data cleaning and manipulating

### 1. cleaning the 'application' data
Firstly,calculate columnwise percentage of null values and drop all those columns which have null values more than 50 percent 
As we can observe that there are neagative values present in columns "DAYS_BIRTH", "DAYS_EMPLOYED", DAYS_REGISTRATION,'DAYS_ID_PUBLISHED'and'DAYS_LAST_PHONE_CHANGE' because they are calculated relative to application so we have to convert them in terms of years by dividing these values by -365 and rounding the value to zero decimal.




```python
#percentage of null values in application data 
per_null=application.isnull().sum()/application.shape[0]*100
per_null
```

```python
#percentage of null values greater than 50 percent
per_null50=per_null[per_null>=50]
per_null50
```

```python
# dropping columns with number of null values greater than 50%
application = application.drop(columns=list(per_null50.index))
```

```python
application.shape
```

```python
application.info(verbose=True)
```

#### here we should drop all the coloumns have  number of null values greater than 50% and handling missing values less than 13%

```python
# finding columns upto 13% null values
per13_null=per_null[(per_null<=13) & (per_null>0)]
per13_null.index
```

```python
# converting negative days of age ,employed days,registration timeand id published at time of application to age and empoyment duration in years  
application['age']= application.DAYS_BIRTH.apply(lambda x: round(x/-365,0))
application['employment duration'] = application.DAYS_EMPLOYED.apply(lambda x: round(x/-365,0) if x<0 else 0)
application['time period of publication of ID'] = application.DAYS_ID_PUBLISH.apply(lambda x: round(x/-365,0) if x<0 else 0)
application['TIME PERIOD OF REGISTRATION'] = application.DAYS_REGISTRATION.apply(lambda x: round(x/-365,0) if x<0 else 0)
application['time period of mobile number change'] = application.DAYS_LAST_PHONE_CHANGE.apply(lambda x: round(x/-365,0) if x<0 else 0)

```

```python
#dropping unnecessary related variable
application = application.drop(columns=["DAYS_BIRTH","DAYS_EMPLOYED",'DAYS_REGISTRATION','DAYS_ID_PUBLISH','DAYS_LAST_PHONE_CHANGE'])
```

## Handling Outliers
Here outliers are present and they can be removed by IQR rule.<br>
Firstly find IQR range<br>
>>IQR=Q3-Q1<br>
>>LOWER LIMIT =Q1-1.5*(IQR)<br> 
>>UPPER LIMIT =Q3+1.5*(IQR)<br>

any data point outside this limit is outlier<br>
we can remove outliers by dropping rows from dataframe where outliers are present<br>

# Top 10 correlations

here we need need upper or lower triangular matrix as all other values are reduntant and diagonal values are result of self-correlation that is always one so it will create problem in sorting

```python
# Select upper triangle of correlation matrix to avoid reduntant valuesbor both cases
upper_corr_1 = corr_1.where(np.triu(np.ones(corr_1.shape), k=1).astype(np.bool))

upper_corr_0 = corr_0.where(np.triu(np.ones(corr_1.shape), k=1).astype(np.bool))
```

```python
#unstacking
upper_corr_1=upper_corr_1.unstack().reset_index()

upper_corr_0=upper_corr_0.unstack().reset_index()
```

```python
#naming columns of correlation matrix
upper_corr_1.columns=['VAR1','VAR2','CORRELATION']

upper_corr_0.columns=['VAR1','VAR2','CORRELATION']
```

```python
#drop null values in correlation column
upper_corr_1.dropna(subset=['CORRELATION'],inplace=True)

upper_corr_0.dropna(subset=['CORRELATION'],inplace=True)
```

```python
# getting absolute values of correlation to include  negativly correlated values in sorted values
upper_corr_1['ABSOLUTE CORRELATION']=upper_corr_1['CORRELATION'].abs()
upper_corr_0['ABSOLUTE CORRELATION']=upper_corr_0['CORRELATION'].abs()

#dropping correlation values greater than .95
upper_corr_1['ABSOLUTE CORRELATION']=upper_corr_1['ABSOLUTE CORRELATION'][upper_corr_1['ABSOLUTE CORRELATION']<.95]
upper_corr_0['ABSOLUTE CORRELATION']=upper_corr_0['ABSOLUTE CORRELATION'][upper_corr_0['ABSOLUTE CORRELATION']<.95]
```

```python
#sorting values by absolute correlation

upper_corr_1=upper_corr_1.sort_values(by='ABSOLUTE CORRELATION',ascending=False)
upper_corr_0=upper_corr_0.sort_values(by='ABSOLUTE CORRELATION',ascending=False)
```
